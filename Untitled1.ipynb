```python
# Импорт необходимых библиотек
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.preprocessing import StandardScaler

# Настройка отображения графиков
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# 1. Считывание данных
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/parkinsons/parkinsons.data"
df = pd.read_csv(url)

# Выбор указанных признаков и метки класса
features = {
    'first': 'MDVP:Fhi(Hz)',    # столбец 3
    'second': 'MDVP:Flo(Hz)',   # столбец 4
    'third': 'DFA',             # столбец 20
    'class': 'status'           # столбец 18
}

data = df[[features['first'], features['second'], features['third'], features['class']]].copy()

print("Первые 5 строк данных:")
print(data.head())
print(f"\nРазмерность данных: {data.shape}")
print(f"\nИнформация о данных:")
print(data.info())

# 2. Обработка пропущенных значений и выбросов
print("\n=== ОБРАБОТКА ПРОПУЩЕННЫХ ЗНАЧЕНИЙ И ВЫБРОСОВ ===")

# Проверка пропущенных значений
missing_class = data[features['class']].isnull().sum()
missing_features = data[features['first']].isnull().sum() + data[features['second']].isnull().sum() + data[features['third']].isnull().sum()

print(f"Пропущенных значений в метках класса: {missing_class}")
print(f"Пропущенных значений в признаках: {missing_features}")

if missing_class > 0:
    data = data.dropna(subset=[features['class']])
    print(f"Удалено записей с пропущенными метками класса: {missing_class}")

if missing_features > 0:
    missing_percentage = (missing_features / len(data)) * 100
    print(f"Процент записей с пропущенными значениями в признаках: {missing_percentage:.2f}%")

    # Замена пропущенных значений на медиану
    for feature in [features['first'], features['second'], features['third']]:
        data[feature] = data[feature].fillna(data[feature].median())
    print("Пропущенные значения заменены на медиану")
else:
    # Удаление выбросов с помощью z-score
    print("Пропущенных значений нет. Удаление выбросов...")

    original_size = len(data)
    z_scores = np.abs(stats.zscore(data[[features['first'], features['second'], features['third']]]))
    outliers = (z_scores > 3).any(axis=1)
    data = data[~outliers]

    removed_percentage = ((original_size - len(data)) / original_size) * 100
    print(f"Удалено записей как выбросы: {original_size - len(data)} ({removed_percentage:.2f}%)")

print(f"Размерность данных после обработки: {data.shape}")

# Анализ данных для последующих шагов
correlation_matrix = data[[features['first'], features['second'], features['third']]].corr()
variance_values = data[[features['first'], features['second'], features['third']]].var()

print(f"\nМатрица корреляции:\n{correlation_matrix}")
print(f"\nДисперсии признаков:\n{variance_values}")

# Определение признаков с наибольшей/наименьшей корреляцией и дисперсией
corr_values = correlation_matrix.abs()
np.fill_diagonal(corr_values.values, 0)  # Исключаем диагональ

max_corr_features = []
for i in range(2):
    max_idx = np.unravel_index(np.argmax(corr_values.values), corr_values.shape)
    max_corr_features = [corr_values.columns[max_idx[0]], corr_values.columns[max_idx[1]]]
    corr_values.iloc[max_idx] = 0

high_var_feature = variance_values.idxmax()
low_var_feature = variance_values.idxmin()

print(f"\nПризнаки с наибольшей корреляцией: {max_corr_features}")
print(f"Признак с наибольшей дисперсией: {high_var_feature}")
print(f"Признак с наименьшей дисперсией: {low_var_feature}")

# 3. Диаграмма рассеяния с наибольшей корреляцией
print("\n=== ДИАГРАММА РАССЕЯНИЯ ===")
plt.figure(figsize=(12, 8))

# Создание маркеров разного размера на основе третьего признака
third_feature = [f for f in [features['first'], features['second'], features['third']] if f not in max_corr_features][0]
sizes = (data[third_feature] - data[third_feature].min()) / (data[third_feature].max() - data[third_feature].min()) * 100 + 10

scatter = sns.scatterplot(
    data=data,
    x=max_corr_features[0],
    y=max_corr_features[1],
    hue=features['class'],
    style=features['class'],
    size=sizes,
    sizes=(20, 200),
    alpha=0.7
)

plt.title(f'Диаграмма рассеяния: {max_corr_features[0]} vs {max_corr_features[1]}', fontsize=14, fontweight='bold')
plt.xlabel(max_corr_features[0], fontsize=12)
plt.ylabel(max_corr_features[1], fontsize=12)

# Создание кастомной легенды для размеров
handles, labels = scatter.get_legend_handles_labels()
third_feature_legend = plt.legend(
    handles[-3:],
    [f'Малый {third_feature}', f'Средний {third_feature}', f'Большой {third_feature}'],
    title=third_feature,
    loc='upper right',
    bbox_to_anchor=(1, 1)
)
plt.gca().add_artist(third_feature_legend)

plt.tight_layout()
plt.show()

# 4. Линейные графики зависимости признаков
print("\n=== ЛИНЕЙНЫЕ ГРАФИКИ ===")
plt.figure(figsize=(12, 8))

lineplot = sns.lineplot(
    data=data,
    x=low_var_feature,
    y=high_var_feature,
    hue=features['class'],
    size=features['class'],
    sizes={0: 2, 1: 4},  # Разный размер для разных классов
    style=features['class'],
    markers=True,
    dashes=False
)

plt.title(f'Зависимость {high_var_feature} от {low_var_feature}', fontsize=14, fontweight='bold')
plt.xlabel(low_var_feature, fontsize=12)
plt.ylabel(high_var_feature, fontsize=12)
plt.legend(title='Status')
plt.tight_layout()
plt.show()

# 5. Диаграмма категориального рассеяния (swarmplot)
print("\n=== SWARMPLOT ===")
plt.figure(figsize=(10, 6))

swarm = sns.swarmplot(
    data=data,
    x=features['class'],
    y=low_var_feature,
    hue=features['class'],
    palette='viridis'
)

plt.title(f'Распределение {low_var_feature} по классам', fontsize=14, fontweight='bold')
plt.xlabel('Status', fontsize=12)
plt.ylabel(low_var_feature, fontsize=12)
plt.legend(title='Status')
plt.tight_layout()
plt.show()

# 6. Диаграмма размаха (boxplot)
print("\n=== BOXPLOT ===")
plt.figure(figsize=(10, 6))

box = sns.boxplot(
    data=data,
    x=features['class'],
    y=high_var_feature,
    hue=features['class'],
    palette='Set2'
)

plt.title(f'Распределение {high_var_feature} по классам', fontsize=14, fontweight='bold')
plt.xlabel('Status', fontsize=12)
plt.ylabel(high_var_feature, fontsize=12)
plt.legend(title='Status')
plt.tight_layout()
plt.show()

# Анализ выбросов для boxplot
print("\nАнализ выбросов:")
Q1 = data[high_var_feature].quantile(0.25)
Q3 = data[high_var_feature].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_mask = (data[high_var_feature] < lower_bound) | (data[high_var_feature] > upper_bound)
outliers_count = outliers_mask.sum()

print(f"В признаке {high_var_feature} обнаружено выбросов: {outliers_count}")

# 7. Столбчатая диаграмма (barplot)
print("\n=== BARPLOT ===")

# Расчет межквартильного размаха для всех признаков
iqr_values = {}
for feature in [features['first'], features['second'], features['third']]:
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    iqr_values[feature] = Q3 - Q1

# Признак с наименьшей разностью между квартилями
min_iqr_feature = min(iqr_values, key=iqr_values.get)
print(f"Признак с наименьшим IQR: {min_iqr_feature} (IQR = {iqr_values[min_iqr_feature]:.4f})")

plt.figure(figsize=(10, 6))

bar = sns.barplot(
    data=data,
    x=features['class'],
    y=min_iqr_feature,
    hue=features['class'],
    estimator='mean',
    errorbar=None,
    palette='pastel'
)

plt.title(f'Среднее значение {min_iqr_feature} по классам', fontsize=14, fontweight='bold')
plt.xlabel('Status', fontsize=12)
plt.ylabel(f'Среднее {min_iqr_feature}', fontsize=12)
plt.legend(title='Status')
plt.tight_layout()
plt.show()

# 8. Диаграмма одномерной оценки плотности
print("\n=== KDE PLOT ===")
plt.figure(figsize=(10, 6))

kde = sns.kdeplot(
    data=data,
    x=features['second'],
    fill=True,
    color='skyblue',
    alpha=0.7
)

plt.title(f'Распределение признака {features["second"]}', fontsize=14, fontweight='bold')
plt.xlabel(features['second'], fontsize=12)
plt.ylabel('Плотность', fontsize=12)
plt.tight_layout()
plt.show()

# 9. Диаграмма двумерной оценки плотности
print("\n=== JOINT PLOT ===")
joint = sns.jointplot(
    data=data,
    x=features['first'],
    y=features['third'],
    kind='kde',
    fill=True,
    height=8
)

joint.fig.suptitle(f'Совместное распределение {features["first"]} и {features["third"]}',
                   fontsize=14, fontweight='bold', y=1.02)
joint.set_axis_labels(features['first'], features['third'], fontsize=12)
plt.tight_layout()
plt.show()

# 10. График регрессии (полиномиальная модель 2-го порядка)
print("\n=== РЕГРЕССИОННЫЙ ГРАФИК ===")
plt.figure(figsize=(10, 6))

reg = sns.regplot(
    data=data,
    x=features['first'],
    y=features['second'],
    order=2,  # Полином второй степени
    scatter_kws={'alpha':0.5},
    line_kws={'color': 'red'}
)

plt.title(f'Регрессия {features["second"]} от {features["first"]}\n(полином 2-й степени)',
          fontsize=14, fontweight='bold')
plt.xlabel(features['first'], fontsize=12)
plt.ylabel(features['second'], fontsize=12)
plt.tight_layout()
plt.show()

print("\n=== ЛАБОРАТОРНАЯ РАБОТА ЗАВЕРШЕНА ===")
```
